## Step 2: Import Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set theme
sns.set(style="whitegrid")
```

## Step 3: Load Dataset


```python
# Load dataset
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Show top 5 rows
df.head()
```

## Step 4: Understand the Data

### 4.1 Shape and Basic Info


```python
df.shape
df.info()
```

### 4.2 Statistical Summary


```python
df.describe()
```

### 4.3 Check Missing Values


```python
df.isnull().sum()
```

## Step 5: Data Cleaning

### 5.1 Convert Columns to Proper Data Types


```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

### 5.2 Drop Missing Values


```python
df.dropna(inplace=True)
```

### 5.3 Final Null Check


```python
df.isnull().sum()
```

## Step 6: Exploratory Data Analysis (EDA)


### 6.1 Churn Count


```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Count")
plt.show()
```

### 6.2 Churn by Contract Type


```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.show()
```

### 6.3 Churn by Monthly Charges


```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges by Churn")
plt.show()
```

### 6.4 Churn by Tenure


```python
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', kde=True)
plt.title("Tenure Distribution by Churn")
plt.show()
```

### 6.5 Churn by Tech Support


```python
sns.countplot(x='TechSupport', hue='Churn', data=df)
plt.title("Churn by Tech Support")
plt.xticks(rotation=45)
plt.show()
```

## Step 7: Key Insights

- Overall churn rate is approximately 26%
- Month-to-month contracts have the highest churn
- Higher monthly charges â†’ more likely to churn
- Short-tenure customers churn more
- Customers with tech support churn less


## Step 8: Conclusion & Recommendations

ðŸ“Œ **Conclusion:**
Our analysis revealed clear patterns in customer churn behavior:
